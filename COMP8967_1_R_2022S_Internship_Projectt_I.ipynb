##Two-class Classification in Scikit-Learn (Case study: k-NN)

K-NN is a method that simply observes which data is lies nearest to the sample of interest which is going to be predicted.  It then classifies the point of interest based on the major class around it. 

```
import pandas as pd

df = pd.read_csv("../content/sample_data/knnData.csv")

df.head()
```

###Normalizing and Feature scaling
All features should be normalized so that each feature contributes eqally to the final distance.

The MinMaxScaler is the most famous scaling algorithm. 
It essentially shrinks the range to 0 to 1 (or -1 to 1 if there are negative values).

This scaler works better for cases in which the standard scaler might not work so well. If the distribution is not Gaussian or the standard deviation is very small, the min-max scaler works better.


Refer to [here](http://benalexkeen.com/feature-scaling-with-scikit-learn/#:~:text=%20Feature%20Scaling%20with%20scikit-learn%20%201%20Standard,in%20n%20-dimensional%20space%20for%20n...%20More) and [here](https://scikit-learn.org/stable/modules/classes.html#module-sklearn.preprocessing)


```
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (4,4))
ax = fig.gca()
df.hist(ax=ax)
plt.show()
```

```
from sklearn import preprocessing

s = preprocessing.MinMaxScaler()
s.fit(X)
X_norm = s.transform(X)

# X = preprocessing.normalize(X, norm='l2')

print(type(X_norm))
```

###Learn k-NN model and fit on data

[k-NN in scikit learn](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html)

[Distance metrics](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.DistanceMetric.html#sklearn.neighbors.DistanceMetric)



```
from sklearn.neighbors import KNeighborsClassifier

k = 6;
# k  actually is the number of neighbors to be considered to measures the distance of test (new) points to nearest neighbors in data
# or, finds the K-nearest neighbors of a point

clf = KNeighborsClassifier(n_neighbors=k , metric = 'euclidean')
# selects the k-nearest data points, and based on those nearest points, the new data point will be assigned the class of majority data points.
knnmodel = clf.fit(X_train, y_train)
```

To estimate the performance for different values of 𝑘 to find the best 𝑘 for dataset, we need to iteratively run the model with different 𝑘 values (with k in range 1 to 20, for example) and finally chose the corresponding 𝑘 that gives higher accuracy.

###Evaluation using [cross validation](https://scikit-learn.org/stable/modules/cross_validation.html) 

* To prevent overfitting
* without directly splitting data to train and validation



**Cross validation**

Learning the parameters of a prediction function and testing it on the same data is a methodological mistake: a model that would just repeat the labels of the samples that it has just seen would have a perfect score but would fail to predict anything useful on yet-unseen data. This situation is called **overfitting**. To avoid it, it is common to use cross validation approach.

For more Study
[Cross validation on scikit-learn](https://scikit-learn.org/stable/modules/cross_validation.html)


```
from sklearn.model_selection import cross_val_score

scores = cross_val_score(clf, X, y, cv=5)

print('Accuracy for each fold is:' , scores)
```

```
print('Average accuracy is:' , scores.mean())
```

By default, the score computed at each CV iteration is the accuracy of the estimator. It is possible to change this by using the [scoring parameter](https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter)


```
from sklearn import metrics

scores = cross_val_score(clf, X, y, cv=5, scoring='precision')
print('Average precision is:' , scores.mean())
```

####Overfitting in Decision Trees, Pruning and Cross Validation

Pruning involves the selective removal of certain parts of a tree to improve the tree’s structure. So that it can adapt if we feed any new kind of data to it, thereby fixing the problem of overfitting.

It reduces the size of a Decision Tree which might slightly increase your training error but drastically decrease your testing error, hence making it more adaptable.


1- Pre-Pruning:

This technique is used before construction of decision tree.

Pre-Pruning can be done using Hyperparameter tuning.

[Decision tree parameters settings](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html#sklearn.tree.DecisionTreeClassifier)

We can say that our model is Overfitting i.e. memorizing the train part but is not able to perform equally well on the test part.

#### ROC Curve

ROC is a metric to evaluate classifier output quality. 
ROC curves typically show a tradeoff between true positive rate on the Y axis, and false positive rate on the X axis. It is ideal to maximize the true positive rate while minimizing the false positive rate. It does mean that a larger area under the curve (AUC) is usually better. The [auc() function from the sklearn.metrics package](https://scikit-learn.org/stable/auto_examples/model_selection/plot_roc.html#sphx-glr-auto-examples-model-selection-plot-roc-py) can be used to produce ROC curves.

```
from sklearn.metrics import plot_roc_curve

ax = plt.subplot()
n = X2.shape[0]
mean_fpr = np.linspace(0, 1, n)
tprs = []
aucs = []
viz = plot_roc_curve(dtModelPruned, X_test, y_test,
                         name='DT',
                         alpha=.3, lw=2, ax=ax)
interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
interp_tpr[0] = 0.0
tprs.append(interp_tpr)
aucs.append(viz.roc_auc)

ax.plot([0, 1], [0, 1], linestyle='--', lw=2, color='r',
label='Chance', alpha=.8)
ax.set(xlim=[-0.05, 1.05], ylim=[-0.05, 1.05],
title="ROC curve comparison")
ax.legend(loc="lower right")
plt.show()
```

Waikato Environment for Knowledge Analysis (**Weka**), developed at the University of Waikato, New Zealand, is an open source software with a complete collection of machine learning algorithms for solving real-world data mining problems. WEKA provides different algorithms for data preprocessing, classification, clustering, feature selection, visualization, and more.
